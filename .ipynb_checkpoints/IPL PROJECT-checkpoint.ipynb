```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sea
```

```python
#load a ipl dataset
ipl = pd.read_csv(r'matches.csv')
ipl
```

```python
#TO SEE FIRST 5 ROWS OF DATA
ipl.head()
```

```python
#TO KNOW THE NUMBER OF ROWS AND COLUMNS
ipl.shape
```

```python
#TO KNOW THE FREQUENCY OF MAN OF THE MATCHES
ipl['player_of_match'].value_counts()
```

```python
#TO KNOW 10 HIGHEST NUMBER OF MAN OF THE MATCHES WINNERS
ipl['player_of_match'].value_counts()[0:10]
```

```python
#TO KNOW TOP 5 MAN OF THE MATCH WINNERS
ipl['player_of_match'].value_counts()[0:5]
```

```python
ll=list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#GETTING A BAR-PLOT OF MOST PLAYER OF THE MATCH WINNERS
plt.figure(figsize=(10,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),ipl['player_of_match'].value_counts()[0:5])
plt.title("MOST MAN OF THE MATCH WINNERS")
plt.xlabel('PLAYERS')
plt.ylabel('Numbers')
                                                                                                    
```

```python
#GETTING FREQUENCY OF RESULT 
ipl['result'].value_counts()
```

```python
#FINDING TOSS WINS W.R.T TEAMS
ipl['toss_winner'].value_counts()
```

```python
#GETTING DATA OF TEAMS WHO WON BY BATTING FIRST
bat_first = ipl[ipl['win_by_runs']!=0]
bat_first
```

```python
bat_first.head()
```

```python
#MAKING A HISTOGRAM FOR RUNS WIN BY THE TEAMS
plt.figure(figsize=(10,5))
plt.hist(bat_first['win_by_runs'],bins=7)
```

```python
#FREQUENCY OF TEAMS WIN BY BATTING FIRST
bat_first['winner'].value_counts()
```

```python
#TOP 3 TEAMS MOST WINS BY BATTING FIRST
bat_first['winner'].value_counts()[0:3]
```

```python
#MAKING A BAR-PLOT OF TOP 3 TEAMS WIN BY BATTING FIRST
plt.figure(figsize=(10,5))
plt.bar(list(bat_first['winner'].value_counts()[0:3].keys()),bat_first['winner'].value_counts()[0:3],color=['blue','yellow','red'])
plt.title('WIN BY BATTING FIRST')
plt.xlabel('TEAMS')
plt.ylabel('NUMBERS')
```

```python
#SHOWING BATTING FIRST PERCENT USING PIE-CHART
plt.figure(figsize = (10,10))
plt.pie(bat_first['winner'].value_counts(),labels = list(bat_first['winner'].value_counts().keys()),autopct='%.1f%%')
```

```python
#TEAMS WON BY BATTING SECOND
bat_second = ipl[ipl['win_by_wickets']!=0]
bat_second
```

```python
bat_second.head()
```

```python
#MAKING A HISTOGRAM FOR THE TEAMS WIN BY SECOND BATTING W.R.T WICKETS
plt.figure(figsize = (10,5))
plt.hist(bat_second['win_by_wickets'],color = 'y',bins = 10)
plt.title('Teams win by wickets')
plt.xlabel('WICKETS')
plt.grid(True)
```

```python
#TEAMS WHO HAVE WON MOST NUMBER OF MATCHES BATTING SECOND
bat_second['winner'].value_counts()
```

```python
bat_second['winner'].value_counts()[0:5]
```

```python
#MAKING A PIE-CHART OF TEAMS WINNING BATTING SECOND
plt.figure(figsize = (10,10))
plt.pie(bat_second['winner'].value_counts(),labels = list(bat_second['winner'].value_counts().keys()),autopct = '%.1f%%')
```

```python
plt.figure(figsize = (12,9))
plt.pie(ipl['winner'].value_counts(),labels = list(ipl['winner'].value_counts().keys()),autopct = '%.1f%%')
```

```python
#A BAR-PLOT OF TOP 3 TEAMS WINNING BATTING SECOND
plt.figure(figsize = (10,6))
plt.bar(list(bat_second['winner'].value_counts()[0:3].keys()),bat_second['winner'].value_counts()[0:3],color=['purple','blue','red'])
```
